```python
import pandas as pd
as_ = pd.read_csv('/content/drive/MyDrive/Diabetes/diabetes.csv') #membac adata  dari file CSV
```

```python
as_.head() # Code ini di gunakan untuk menampilkan 5 baris pertama
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for Outcome, d in as_.groupby('Outcome'):
    ax.scatter(d['Glucose'], d['BloodPressure'], label=Outcome)

plt.legend()
plt.title('Diabetes')
plt.xlabel('Glucose')##  untuk membuat scatter plot yang memvisualisasikan hubungan antara 'Glucose' dan 'BloodPressure' berdasarkan 'Outcome' dari dataset diabetes.
plt.ylabel('BloodPressure')
plt.grid(True)
plt.show()
```

```python
import numpy as np

x_train = np.array(as_[['Glucose', 'BloodPressure']])
y_train = np.array(as_['Outcome']) ## Kode ini bertujuan untuk menyiapkan data untuk pelatihan model machine learning. Data dipisahkan menjadi fitur (X) dan target (y).

##Berikut penjelasan fungsi dari setiap baris kode:

print(f'x_train:\n{x_train}\n')
print(f'y_train: {y_train}')
```

```python
from sklearn.preprocessing  import LabelBinarizer

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(f'y_train:\n{y_train}\n')

##Kode ini bertujuan untuk melakukan transformasi label pada data target y_train menggunakan LabelBinarizer dari library scikit-learn.
```

```python
y_train = y_train.flatten()
print(f'y_train: {y_train}') ##Kode ini bertujuan untuk mengubah bentuk array y_train menjadi satu dimensi menggunakan method flatten().
```

```python
from sklearn.neighbors import KNeighborsClassifier

k = 3
model = KNeighborsClassifier(n_neighbors=k)
model.fit(x_train, y_train)  ##Kode ini bertujuan untuk membuat dan melatih model K-Nearest Neighbors (KNN) menggunakan library scikit-learn
```

```python
Glucose = 148
BloodPressure = 72

x_new = np.array([Glucose, BloodPressure]).reshape(1, -1)
x_new ## Kode ini bertujuan untuk membuat array NumPy x_new yang berisi data baru yang ingin diprediksi oleh model KNN. Data baru ini dibentuk agar sesuai dengan format yang diharapkan oleh model.
```

```python
y_new = model.predict(x_new)
y_new ##Kode ini bertujuan untuk melakukan prediksi menggunakan model KNN yang telah dilatih sebelumnya pada data baru x_new.
```

```python
lb.inverse_transform(y_new) ##Kode ini bertujuan untuk mengubah hasil prediksi y_new yang berupa representasi biner (0 dan 1) kembali ke label aslinya menggunakan method inverse_transform() dari objek LabelBinarizer (lb).
```

```python
fig, ax = plt.subplots()
for Outcome, d in as_.groupby('Outcome'):
    ax.scatter(d['Glucose'], d['BloodPressure'], label=Outcome)

plt.scatter(Glucose,
            BloodPressure,
            color='red',
            marker='s',
            label = 'misterius')

plt.legend(loc='upper left')
plt.title('Diabetes')
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')
plt.grid(True)
plt.show()
plt.show()    ## kode ini bertujuan untuk memvisualisasikan data diabetes berdasarkan 'Glucose' dan 'BloodPressure',
    ##mengelompokkannya berdasarkan 'Outcome', dan menandai data 'misterius' pada plot. Ini membantu dalam memahami distribusi data dan hubungan antar variabel.
```

```python
misterius = np.array([Glucose, BloodPressure])
misterius ##Kode ini bertujuan untuk membuat array NumPy misterius yang berisi data baru yang ingin diprediksi, yaitu nilai Glucose dan BloodPressure.
```

```python
from scipy.spatial.distance import euclidean

data_jarak = [euclidean(misterius, d) for d in x_train]
data_jarak ##Kode ini bertujuan untuk menghitung jarak Euclidean antara data baru (misterius) dengan setiap data point dalam data training (x_train) dan menyimpan hasilnya dalam list data_jarak.
```

```python
as_['jarak'] = data_jarak
as_.sort_values('jarak') ##Kode ini bertujuan untuk menambahkan kolom baru bernama 'jarak' ke DataFrame as_ yang berisi jarak Euclidean yang telah dihitung sebelumnya, dan kemudian mengurutkan DataFrame berdasarkan kolom 'jarak'.
```

```python
x_test = np.array([[148 , 72], [183, 84], [85, 66], [89, 66]])
y_test = lb.transform(np.array(['1', '1', '0', '0']))

print(f'x_test:\n{x_test}\n')
print(f'y_test:\n{y_test}') ##x_test = np.array([[148 , 72], [183, 84], [85, 66], [89, 66]])
y_test = lb.transform(np.array(['1', '1', '0', '0']))

print(f'x_test:\n{x_test}\n')
print(f'y_test:\n{y_test}')
```

```python
y_pred = model.predict(x_test)
y_pred ##Kode ini bertujuan untuk melakukan prediksi pada data testing (x_test) menggunakan model KNN yang telah dilatih sebelumnya dan menyimpan hasil prediksi dalam variabel y_pred.
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print(f'Accuracy: {acc}')
```

```python
from sklearn.metrics import precision_score
prec = precision_score(y_test, y_pred)
print(f'Precision: {prec}') ##Kode ini bertujuan untuk menghitung precision dari model KNN berdasarkan hasil prediksi pada data testing (y_pred) dan label target yang sebenarnya (y_test).
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred)
print(f'Recall:{recall}') ##Kode ini bertujuan untuk menghitung recall dari model KNN berdasarkan hasil prediksi pada data testing (y_pred) dan label target yang sebenarnya (y_test).
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print(f'F1 Score: {f1}') ##Kode tersebut bertujuan untuk menghitung skor F1 dari model KNN berdasarkan hasil prediksi pada data testing (y_pred) dan label target yang sebenarnya (y_test).
```

```python
from sklearn.metrics import classification_report
cls_report = classification_report(y_test, y_pred)
print(f'Classification Report:\n{cls_report}') ## Kode ini bertujuan untuk menghasilkan dan menampilkan classification report dari model KNN berdasarkan hasil prediksi pada data testing (y_pred) dan label target yang sebenarnya (y_test).
```
